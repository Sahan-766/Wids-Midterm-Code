```python
import numpy as np
a =np.random.randint(2,100,(5,5))
print(a)
```

```python
#(a) middle element of the array
middle_element=a[2,2]
print(middle_element)
```

```python
#(b) mean of each row of the array
row_mean=np.mean(a,axis=1)
print(row_mean)
```

```python
#(c) original array that are greater than the overall mean of the array
new_array=a[a>np.mean(a)]
print(new_array)
```

```python
import numpy as np

def numpy_spiral_order(matrix):
    if matrix.size == 0:
        return []
    if matrix.shape[0] == 1:
        return matrix[0].tolist()
    if matrix.shape[1] == 1:
        return matrix[:, 0].tolist()

    spiral = []
    spiral.extend(matrix[0, :].tolist())
    spiral.extend(matrix[1:-1, -1].tolist())
    spiral.extend(matrix[-1, ::-1].tolist())
    spiral.extend(matrix[-2:0:-1, 0].tolist())
    spiral.extend(numpy_spiral_order(matrix[1:-1, 1:-1]))

    return spiral
```

```python
# Testing the spiral function
t=np.arange(1,26,1)
t=np.reshape(t,(5,5))
print(t)
print(numpy_spiral_order(t))

```
