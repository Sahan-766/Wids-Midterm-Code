## Series

```python
# one-dimensional labeled array in Python, capable of holding any data type (integers, strings, floats, etc.).
# It is part of the Pandas library and is similar to a column in a spreadsheet or a single list in Python.

# list to series
l=[1,2,3,4,5]
s=pd.Series(l) # indexing starts from 0
s


l=[1,2.3,'mahan'] # different datatypes
s=pd.Series(l)
print(type(s)) #<class 'pandas.core.series.Series'>

# np array to series
a=np.array([1,2,3,4,5])
s=pd.Series(a)
s


# custom indexing
s=pd.Series(data=[1,2,3,4,5],index=[1,2,3,4,5])
s


# dictionary to series
d={'a':1,'b':2,'c':3,'d':4,'e':5}
s=pd.Series(d) # dictionary keys become the indices and the values become the data in the Series.
s
```

```python
# repeat function

pd.Series(5)
pd.Series(5).repeat(5)
pd.Series(5).repeat(5).reset_index(drop=True) #The drop=True parameter ensures that the old index is not added as a column.

# otherwise it will create a df
pd.Series(5).repeat(5).reset_index()
```

```python
# indexing, slicing

s=pd.Series([1,2]).repeat([5,2]).reset_index(drop=True) # repeat 1 and 2 - 5 and 2 times resp.
print(s)
s[0]
s[6]
s[0:3]
s[0:5:2]
s[::-1]
s[:5]
s[3:]
s[1:-1] # removes first and last elements
s[2:-2] # removes first and last 2 elements
```

```python
# Aggregate function

s=pd.Series([1,2,3,4,5])
s.max()
s.min()
s.sum()
s.mean()
s.var()
s.std()

s.agg([min,max,sum])
```

```python
# absolute function

s=pd.Series([1,2,-1,-2,3,4,-4])
s
s.abs() # -ve to +ve
```

```python
# Appending Series

s1=pd.Series([1,2,3,4])
s2=pd.Series([5,6,7,8])
print(s1)
print(s2)

s1.append(s2) # indeces remain same
s2.append(s1)

s1.append(s2).reset_index(drop=True) # to get consecutive indeces
```

```python
# astype fun

print(s1) # int
s1.astype('float') # converts to float
s1.astype('str')
```

```python
# between function

s=pd.Series([1,2,10,4,5,6,7,8])
print(s)
s.between(7,10) # returns bool type series with mentioned interval values are true remaining false
```

```python
# string functions

s=pd.Series(['mah an ',' SAHAN','Abhinay','GOWtham'])
print(s)

s.str.upper()
s.str.lower()
s.str.len()

for i in s:
    print(len(i))


#strip function
s=s.str.strip() # It removes any whitespace (spaces, tabs, newlines) from the beginning and end of each string, leaving the middle spaces (if any) intact.
for i in s:
    print(len(i))


# split function
s=pd.Series(['mah an ',' SAHAN','Abhinay','GOWtham'])
print("After split: \n",s.str.split()) # each str element gets split based on white spaces resulting in lists of the split components.


# can be split on other factors too (default is whitespace)
s=pd.Series(['ma/ha/n','sa/han'])
s.str.split('/')
s.str.split('/')[0]
s.str.split('/')[0][0]


# contains function
s=pd.Series(['mah an ',' SAHAN','Abhinay','GOWtham'])
s.str.contains('m') # true or false


# replace
s.str.replace('m','MM')

# count
s.str.count('m')

# startswith and endswith
s.str.endswith('N')
s.str.startswith(' ')

# find function - The s.str.find() method in Pandas is used to find the position (index) of the first occurrence of a substring within each string element of a Series. If the substring is not found, it returns -1.
s = pd.Series(['apple', 'banana', 'cherry', 'date'])
print(s)
s.str.find('an')
```

```python
# series to list
print(s)
s.tolist()
```

```python
# slicing using loc and iloc

data = {'ID': [101, 102, 103], 'Name': ['John', 'Alice', 'Bob'], 'Age': [25, 30, 22]}
df = pd.DataFrame(data, index=['a', 'b', 'c'])
df

# 1.  loc

#: Label-based indexing (allows you to select rows and columns based on explicit labels(index name or column name))
row_a=df.loc['a'] # selects the row with index label 'a'

# Selecting multiple rows and specific columns
subset = df.loc[['a', 'b'], ['Name', 'Age']]
print(subset)
# print(type(subset)) - dataframe

# Accessing the 'Name' column using loc
df.loc[:, 'Name']

# Accessing the 'Name' and 'Age' columns
name_age_columns = df.loc[:, ['Name', 'Age']]
print(name_age_columns)


# ------------------------------------------------

# 2. iloc

# Position-based indexing: accessing data by the integer position (row or column index).
# access rows and columns by their numerical index positions
#Endpoints are exclusive
row_0 = df.iloc[0]
print(row_0)

# Selecting specific rows and columns by position
subset = df.iloc[0:2, 1:3]  # First two rows, columns 1 and 2
print(subset)


print(df.iloc[1:3]) # all cols, rows- 2,3
```

```python
# slicing using conditions

d1=df.loc[:,['Age']]
d2=df['Age']>24
df[d2]
```

```python
# column addition

data = {'ID': [101, 102, 103], 'Name': ['John', 'Alice', 'Bob'], 'Age': [25, 30, 22]}
df = pd.DataFrame(data, index=['a', 'b', 'c'])
df

# using list
df['newcol']=[1,2,3]
df

# using series
s=pd.Series([1,2,3],index=['a','b','c']) # to match the indeces of existing df
df['newCol2']=s
df


# add new col based on existing col
df['ID+']=df['ID']+1
df
```

```python
#  Statistical Functions

d={'Ones':[1,2,3,4,5],'Tens':[10,20,30,40,50],'Hundreds':[100,200,300,400,500],'Thousands':[1000,2000,3000,4000,5000]}
df=pd.DataFrame(d)
df

df.sum() # sum of each col
df.mean() # of each col
df.median() # of each col

df['Ones'].mode()
df['Tens'].mode()
df.mode() # mode of each col

df.var() # gives series
df.std()
df.min()
df.max()

df.describe() # gives all stats of each column
# print(type(df.describe())) --> dataframe
```

```python
# pipe function

data={'one':pd.Series([1,2,3,4]),
      'two':pd.Series([10,20,30,40]),
      'three':pd.Series([100,200,300,400]),
      'four':pd.Series([1000,2000,3000,4000])}
df=pd.DataFrame(data)
df

def add_(a,b):
    return a+b

def sub_(a,b):
    return a-b

def mean_(col):
    return col.mean()

def square_(a):
    return a*a


df.pipe(add_,10) # adds 10 to all cells. (here parameter a is the value in the cell and b is 10 fixed)
df.pipe(sub_,20) # subs 10 from all cells
df.pipe(mean_) # mean of each col
df.pipe(square_) # square of all cells

df.pipe(mean_).pipe(square_)
```

```python
# Apply function -axis wise

print(df)

df.apply(np.mean) # mean of each col
df.apply(np.min) # min of each col

df.apply(lambda x:x.max()-x.min()) # for each column
```

```python
# Apply map function - elementwise

print(df)
df.applymap(lambda X:X*100) # applies for each element
```

```python
# Reindex function - used to change row lables or column labels

print(df)

df=df.reindex([1,2,0,3]) # or use inplace=True
print('\n',df) # rows order changes

df= df.reindex(columns=['two','one','four','three']) # for columns
print('\n',df)
```

```python
# renaming columns

data={'one':pd.Series([1,2,3,4]),
      'two':pd.Series([10,20,30,40]),
      'three':pd.Series([100,200,300,400]),
      'four':pd.Series([1000,2000,3000,4000])}
df=pd.DataFrame(data)
df

df.rename(columns={'one':'ONE','two':'TWO'},inplace=True)
df

# to rename rows
df.rename(index={0:'a',1:'b',2:'c',3:'d'},inplace=True)
df
```

```python
# sorting by column (similar to order by clause in SQL)

data={'one':pd.Series([1,1,3,2]),
      'two':pd.Series([40,20,30,10]),
      'three':pd.Series([100,200,300,400]),
      'four':pd.Series([1000,2000,3000,4000])}
df=pd.DataFrame(data)
print(df)

df.sort_values(by='one') # similar to order by clause in sql
df.sort_values(by=['one','two'])

df.sort_values(by='one',ascending=False,kind='mergesort')
```

```python
# groupby function
data={'country':['india','china','usa','japan','india','usa','china','india'],
      'rank':[1,2,3,4,5,8,7,6],
      'year':[2014,2014,2015,2014,2016,2015,2017,2016],
      'points':[100,200,300,100,200,400,200,100]
      }
df=pd.DataFrame(data)
print(df)


df.groupby('country').groups # gives the indeces of rows of each group
df.groupby(['country','year']).groups

df.groupby(['country','year']).get_group(('india',2014))

df.groupby('country')['points'].sum() # total points by each country
df.groupby('country')['points'].sum().sort_values(ascending=False) # sorts in descending order


groups=df.groupby('country')
groups['points'].agg([np.sum,np.mean,np.std,np.max,np.min])
# print(type(groups))

groups=df.groupby('country')
groups.filter(lambda x:len(x)==2) # country group that appear 2 times
```

```python
# dropna function -- drop null values

# SYNTAX: df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)

df.head()
df.isnull().sum() # here there are no null values
df.dropna()  # inplace=True # drops all rows with null


data = {'A': [1, 2, np.nan, 4],
        'B': [5, np.nan, np.nan, 8],
        'C': [9, 10, 11, 12]}
df = pd.DataFrame(data)

df_cleaned = df.dropna() #  Dropping rows with NaN
print(df_cleaned)

df_cleaned = df.dropna(axis=1) #Dropping columns with NaN
print(df_cleaned)
```

```python
# fillna function -ill missing (NaN) values in a DataFrame or Series with a specified value, method, or strategy

# SYNTAX: df.fillna(value=None, method=None, axis=None, inplace=False, limit=None)

df.fillna(0) #This fills all missing values with 0
df.fillna({'A': 0, 'B': 5,'C':'missing'}) # here A,B and c are columns --> different fill values for different columns.
df.fillna(method='ffill') # previous non-null value forward to fill missing data.
df.fillna(method='bfill') # propagating the next non-null value backward.
df.fillna(df.mean(), inplace=True)  # Fills NaNs with the mean of each column

df_filled = df.fillna(method='ffill').fillna(method='bfill')
print("Combined Fill:\n", df_filled)

df['A'].fillna(100) # fills null values of col-A with 100

# for numerical nan-- fill with mean
# for categorical nan-- fill with mode
```
